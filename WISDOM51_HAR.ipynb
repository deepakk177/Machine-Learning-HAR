```python
# Human Activity Recognition using Accelerometer and Gyroscope Data
# Complete ML/DL Pipeline with Classical ML and Deep Learning Models

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.decomposition import PCA
import xgboost as xgb
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Deep Learning imports
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model

# For Kaggle API (uncomment if using Kaggle dataset)
# import kaggle
# import zipfile
# import os

import pickle
import joblib

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class HumanActivityRecognizer:
    def __init__(self):
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        self.models = {}
        self.history = {}

    def load_kaggle_dataset(self, dataset_name="uciml/human-activity-recognition-with-smartphones"):
        """
        Load dataset from Kaggle API
        Uncomment and configure kaggle.json file in ~/.kaggle/ directory
        """
        # kaggle.api.dataset_download_files(dataset_name, path='./data/', unzip=True)
        pass

    def load_data(self, file_path=None, use_sample_data=True):
        """
        Load and prepare the dataset
        If no file_path provided, creates sample data for demonstration
        """
        if use_sample_data or file_path is None:
            print("Creating sample dataset for demonstration...")
            # Create sample data similar to UCI HAR dataset
            n_samples = 10000
            n_features = 6  # 3 accelerometer + 3 gyroscope
            n_timesteps = 128

            # Generate synthetic sensor data
            np.random.seed(42)
            data = []
            labels = []
            activities = ['WALKING', 'WALKING_UPSTAIRS', 'WALKING_DOWNSTAIRS',
                         'SITTING', 'STANDING', 'LAYING']

            for i in range(n_samples):
                activity_idx = np.random.randint(0, 6)
                activity = activities[activity_idx]

                # Generate synthetic accelerometer and gyroscope data
                # Different activities have different patterns
                if activity == 'WALKING':
                    acc_x = np.sin(np.linspace(0, 4*np.pi, n_timesteps)) + np.random.normal(0, 0.1, n_timesteps)
                    acc_y = np.cos(np.linspace(0, 4*np.pi, n_timesteps)) + np.random.normal(0, 0.1, n_timesteps)
                    acc_z = np.ones(n_timesteps) * 9.8 + np.random.normal(0, 0.2, n_timesteps)
                elif activity in ['SITTING', 'STANDING']:
                    acc_x = np.random.normal(0, 0.05, n_timesteps)
                    acc_y = np.random.normal(0, 0.05, n_timesteps)
                    acc_z = np.ones(n_timesteps) * 9.8 + np.random.normal(0, 0.1, n_timesteps)
                elif activity == 'LAYING':
                    acc_x = np.random.normal(0, 0.03, n_timesteps)
                    acc_y = np.ones(n_timesteps) * 9.8 + np.random.normal(0, 0.1, n_timesteps)
                    acc_z = np.random.normal(0, 0.03, n_timesteps)
                else:  # WALKING_UPSTAIRS, WALKING_DOWNSTAIRS
                    acc_x = np.sin(np.linspace(0, 3*np.pi, n_timesteps)) + np.random.normal(0, 0.15, n_timesteps)
                    acc_y = np.cos(np.linspace(0, 3*np.pi, n_timesteps)) + np.random.normal(0, 0.15, n_timesteps)
                    acc_z = np.ones(n_timesteps) * 9.8 + np.random.normal(0, 0.3, n_timesteps)

                # Generate gyroscope data
                gyro_x = np.random.normal(0, 0.1, n_timesteps)
                gyro_y = np.random.normal(0, 0.1, n_timesteps)
                gyro_z = np.random.normal(0, 0.1, n_timesteps)

                # Combine features
                sample = np.column_stack([acc_x, acc_y, acc_z, gyro_x, gyro_y, gyro_z])
                data.append(sample)
                labels.append(activity)

            self.X = np.array(data)
            self.y = np.array(labels)

        else:
            # Load from CSV file
            print(f"Loading data from {file_path}...")
            df = pd.read_csv(file_path)
            # Assume the last column is the target variable
            self.X = df.iloc[:, :-1].values
            self.y = df.iloc[:, -1].values

        print(f"Data shape: {self.X.shape}")
        print(f"Labels shape: {self.y.shape}")
        print(f"Unique activities: {np.unique(self.y)}")

        return self.X, self.y

    def preprocess_data(self, test_size=0.2, val_size=0.1):
        """
        Preprocess the data: encode labels, normalize features, split data
        """
        # Encode labels
        self.y_encoded = self.label_encoder.fit_transform(self.y)
        self.n_classes = len(np.unique(self.y_encoded))

        # Split data
        X_temp, self.X_test, y_temp, self.y_test = train_test_split(
            self.X, self.y_encoded, test_size=test_size, random_state=42, stratify=self.y_encoded
        )

        self.X_train, self.X_val, self.y_train, self.y_val = train_test_split(
            X_temp, y_temp, test_size=val_size/(1-test_size), random_state=42, stratify=y_temp
        )

        print(f"Train shape: {self.X_train.shape}, Val shape: {self.X_val.shape}, Test shape: {self.X_test.shape}")

        return self.X_train, self.X_val, self.X_test, self.y_train, self.y_val, self.y_test

    def extract_features(self, X):
        """
        Extract statistical features from time-series data for classical ML models
        """
        n_samples, n_timesteps, n_features = X.shape
        extracted_features = []

        for i in range(n_samples):
            sample_features = []
            for j in range(n_features):
                signal = X[i, :, j]

                # Statistical features
                sample_features.extend([
                    np.mean(signal),           # Mean
                    np.std(signal),            # Standard deviation
                    np.min(signal),            # Minimum
                    np.max(signal),            # Maximum
                    np.median(signal),         # Median
                    stats.skew(signal),        # Skewness
                    stats.kurtosis(signal),    # Kurtosis
                    np.var(signal),            # Variance
                    np.sqrt(np.mean(signal**2)),  # RMS
                    np.sum(signal**2)          # Energy
                ])

                # Additional features
                sample_features.append(np.percentile(signal, 25))  # 25th percentile
                sample_features.append(np.percentile(signal, 75))  # 75th percentile
                sample_features.append(np.max(signal) - np.min(signal))  # Range

            extracted_features.append(sample_features)

        return np.array(extracted_features)

    def train_classical_models(self, apply_pca=False, n_components=50):
        """
        Train classical ML models (RandomForest, SVM, XGBoost)
        """
        print("Extracting features for classical ML models...")
        X_train_features = self.extract_features(self.X_train)
        X_val_features = self.extract_features(self.X_val)
        X_test_features = self.extract_features(self.X_test)

        # Normalize features
        X_train_scaled = self.scaler.fit_transform(X_train_features)
        X_val_scaled = self.scaler.transform(X_val_features)
        X_test_scaled = self.scaler.transform(X_test_features)

        # Apply PCA if requested
        if apply_pca:
            print(f"Applying PCA with {n_components} components...")
            self.pca = PCA(n_components=n_components)
            X_train_scaled = self.pca.fit_transform(X_train_scaled)
            X_val_scaled = self.pca.transform(X_val_scaled)
            X_test_scaled = self.pca.transform(X_test_scaled)

        # Define models
        models = {
            'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
            'SVM': SVC(kernel='rbf', random_state=42),
            'XGBoost': xgb.XGBClassifier(random_state=42, eval_metric='mlogloss')
        }

        # Train models
        for name, model in models.items():
            print(f"\nTraining {name}...")
            model.fit(X_train_scaled, self.y_train)

            # Evaluate on validation set
            val_pred = model.predict(X_val_scaled)
            val_acc = accuracy_score(self.y_val, val_pred)
            print(f"{name} Validation Accuracy: {val_acc:.4f}")

            self.models[name] = {
                'model': model,
                'val_accuracy': val_acc,
                'X_test': X_test_scaled
            }

        return self.models

    def hyperparameter_tuning(self, model_name='RandomForest'):
        """
        Perform hyperparameter tuning using GridSearchCV
        """
        print(f"Performing hyperparameter tuning for {model_name}...")

        X_train_features = self.extract_features(self.X_train)
        X_train_scaled = self.scaler.fit_transform(X_train_features)

        if model_name == 'RandomForest':
            param_grid = {
                'n_estimators': [50, 100, 200],
                'max_depth': [None, 10, 20],
                'min_samples_split': [2, 5, 10]
            }
            model = RandomForestClassifier(random_state=42)

        elif model_name == 'SVM':
            param_grid = {
                'C': [0.1, 1, 10],
                'gamma': ['scale', 'auto', 0.001, 0.01]
            }
            model = SVC(kernel='rbf', random_state=42)

        grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
        grid_search.fit(X_train_scaled, self.y_train)

        print(f"Best parameters: {grid_search.best_params_}")
        print(f"Best cross-validation score: {grid_search.best_score_:.4f}")

        return grid_search.best_estimator_

    def build_lstm_model(self, input_shape, lstm_units=64, dropout_rate=0.2):
        """
        Build LSTM model for sequence classification
        """
        model = Sequential([
            LSTM(lstm_units, return_sequences=True, input_shape=input_shape),
            Dropout(dropout_rate),
            LSTM(lstm_units//2, return_sequences=False),
            Dropout(dropout_rate),
            Dense(64, activation='relu'),
            Dropout(dropout_rate),
            Dense(self.n_classes, activation='softmax')
        ])

        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def build_gru_model(self, input_shape, gru_units=64, dropout_rate=0.2):
        """
        Build GRU model for sequence classification
        """
        model = Sequential([
            GRU(gru_units, return_sequences=True, input_shape=input_shape),
            Dropout(dropout_rate),
            GRU(gru_units//2, return_sequences=False),
            Dropout(dropout_rate),
            Dense(64, activation='relu'),
            Dropout(dropout_rate),
            Dense(self.n_classes, activation='softmax')
        ])

        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def build_cnn_model(self, input_shape, dropout_rate=0.3):
        """
        Build CNN model for sequence classification
        """
        model = Sequential([
            Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape),
            MaxPooling1D(pool_size=2),
            Dropout(dropout_rate),
            Conv1D(128, kernel_size=3, activation='relu'),
            MaxPooling1D(pool_size=2),
            Dropout(dropout_rate),
            Conv1D(64, kernel_size=3, activation='relu'),
            Flatten(),
            Dense(100, activation='relu'),
            Dropout(dropout_rate),
            Dense(self.n_classes, activation='softmax')
        ])

        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def train_deep_learning_models(self, lstm_units=64, gru_units=64, dropout_rate=0.3,
                                 batch_size=32, epochs=30):
        """
        Train deep learning models (LSTM, GRU, CNN)
        """
        input_shape = (self.X_train.shape[1], self.X_train.shape[2])

        # Normalize data for deep learning
        X_train_norm = self.X_train.reshape(-1, self.X_train.shape[-1])
        X_val_norm = self.X_val.reshape(-1, self.X_val.shape[-1])
        X_test_norm = self.X_test.reshape(-1, self.X_test.shape[-1])

        scaler_dl = StandardScaler()
        X_train_norm = scaler_dl.fit_transform(X_train_norm)
        X_val_norm = scaler_dl.transform(X_val_norm)
        X_test_norm = scaler_dl.transform(X_test_norm)

        X_train_norm = X_train_norm.reshape(self.X_train.shape)
        X_val_norm = X_val_norm.reshape(self.X_val.shape)
        X_test_norm = X_test_norm.reshape(self.X_test.shape)

        self.scaler_dl = scaler_dl

        # Define callbacks
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        dl_models = {
            'LSTM': self.build_lstm_model(input_shape, lstm_units, dropout_rate),
            'GRU': self.build_gru_model(input_shape, gru_units, dropout_rate),
            'CNN': self.build_cnn_model(input_shape, dropout_rate)
        }

        for name, model in dl_models.items():
            print(f"\nTraining {name} model...")

            # Model checkpoint
            checkpoint = ModelCheckpoint(
                f'best_{name.lower()}_model.h5',
                monitor='val_accuracy',
                save_best_only=True,
                mode='max',
                verbose=1
            )

            # Train model
            history = model.fit(
                X_train_norm, self.y_train,
                batch_size=batch_size,
                epochs=epochs,
                validation_data=(X_val_norm, self.y_val),
                callbacks=[checkpoint, early_stopping],
                verbose=1
            )

            # Evaluate on test set
            test_loss, test_acc = model.evaluate(X_test_norm, self.y_test, verbose=0)
            print(f"{name} Test Accuracy: {test_acc:.4f}")

            self.models[name] = {
                'model': model,
                'history': history,
                'test_accuracy': test_acc,
                'X_test': X_test_norm
            }

            self.history[name] = history

        return self.models, self.history

    def plot_training_history(self):
        """
        Plot training and validation loss/accuracy for deep learning models
        """
        fig, axes = plt.subplots(2, 3, figsize=(18, 10))
        fig.suptitle('Deep Learning Models Training History', fontsize=16)

        dl_models = ['LSTM', 'GRU', 'CNN']

        for i, model_name in enumerate(dl_models):
            if model_name in self.history:
                history = self.history[model_name]

                # Plot accuracy
                axes[0, i].plot(history.history['accuracy'], label='Train Accuracy')
                axes[0, i].plot(history.history['val_accuracy'], label='Val Accuracy')
                axes[0, i].set_title(f'{model_name} - Accuracy')
                axes[0, i].set_xlabel('Epoch')
                axes[0, i].set_ylabel('Accuracy')
                axes[0, i].legend()
                axes[0, i].grid(True)

                # Plot loss
                axes[1, i].plot(history.history['loss'], label='Train Loss')
                axes[1, i].plot(history.history['val_loss'], label='Val Loss')
                axes[1, i].set_title(f'{model_name} - Loss')
                axes[1, i].set_xlabel('Epoch')
                axes[1, i].set_ylabel('Loss')
                axes[1, i].legend()
                axes[1, i].grid(True)

        plt.tight_layout()
        plt.show()

    def evaluate_models(self):
        """
        Evaluate all trained models and generate comprehensive reports
        """
        results = {}

        for model_name, model_info in self.models.items():
            print(f"\n{'='*50}")
            print(f"EVALUATION RESULTS FOR {model_name.upper()}")
            print(f"{'='*50}")

            model = model_info['model']
            X_test = model_info['X_test']

            # Make predictions
            if model_name in ['LSTM', 'GRU', 'CNN']:
                y_pred_proba = model.predict(X_test)
                y_pred = np.argmax(y_pred_proba, axis=1)
            else:
                y_pred = model.predict(X_test)
                try:
                    y_pred_proba = model.predict_proba(X_test)
                except:
                    y_pred_proba = None

            # Calculate metrics
            accuracy = accuracy_score(self.y_test, y_pred)

            print(f"Test Accuracy: {accuracy:.4f}")
            print("\nClassification Report:")
            print(classification_report(
                self.y_test, y_pred,
                target_names=self.label_encoder.classes_,
                digits=4
            ))

            # Confusion Matrix
            cm = confusion_matrix(self.y_test, y_pred)

            plt.figure(figsize=(8, 6))
            sns.heatmap(
                cm,
                annot=True,
                fmt='d',
                cmap='Blues',
                xticklabels=self.label_encoder.classes_,
                yticklabels=self.label_encoder.classes_
            )
            plt.title(f'Confusion Matrix - {model_name}')
            plt.xlabel('Predicted')
            plt.ylabel('Actual')
            plt.show()

            results[model_name] = {
                'accuracy': accuracy,
                'predictions': y_pred,
                'confusion_matrix': cm
            }

            if y_pred_proba is not None:
                results[model_name]['probabilities'] = y_pred_proba

        return results

    def cross_validate_classical_models(self, cv_folds=5):
        """
        Perform cross-validation for classical ML models
        """
        print("\nPerforming Cross-Validation for Classical ML Models...")

        X_features = self.extract_features(self.X_train)
        X_scaled = self.scaler.fit_transform(X_features)

        classical_models = ['RandomForest', 'SVM', 'XGBoost']
        cv_results = {}

        for model_name in classical_models:
            if model_name in self.models:
                model = self.models[model_name]['model']
                cv_scores = cross_val_score(model, X_scaled, self.y_train, cv=cv_folds, scoring='accuracy')

                print(f"\n{model_name} Cross-Validation Results:")
                print(f"Mean Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
                print(f"Individual Fold Scores: {cv_scores}")

                cv_results[model_name] = {
                    'mean_accuracy': cv_scores.mean(),
                    'std': cv_scores.std(),
                    'scores': cv_scores
                }

        return cv_results

    def save_models(self):
        """
        Save trained models to disk
        """
        print("Saving models...")

        # Save classical ML models
        for model_name, model_info in self.models.items():
            if model_name in ['RandomForest', 'SVM', 'XGBoost']:
                joblib.dump(model_info['model'], f'{model_name.lower()}_model.pkl')
                print(f"Saved {model_name} model as {model_name.lower()}_model.pkl")

        # Save deep learning models (already saved during training with ModelCheckpoint)
        for model_name in ['LSTM', 'GRU', 'CNN']:
            if model_name in self.models:
                print(f"{model_name} model already saved as best_{model_name.lower()}_model.h5")

        # Save preprocessors
        joblib.dump(self.scaler, 'feature_scaler.pkl')
        joblib.dump(self.label_encoder, 'label_encoder.pkl')
        if hasattr(self, 'scaler_dl'):
            joblib.dump(self.scaler_dl, 'dl_scaler.pkl')

        print("All models and preprocessors saved successfully!")

    def load_saved_model(self, model_name, model_path):
        """
        Load a saved model from disk
        """
        if model_name in ['LSTM', 'GRU', 'CNN']:
            model = load_model(model_path)
        else:
            model = joblib.load(model_path)

        return model

    def predict_single_sample(self, sample_data, model_name):
        """
        Predict activity for a single sample
        """
        if model_name not in self.models:
            raise ValueError(f"Model {model_name} not found. Available models: {list(self.models.keys())}")

        model = self.models[model_name]['model']

        if model_name in ['LSTM', 'GRU', 'CNN']:
            # For deep learning models
            if len(sample_data.shape) == 2:
                sample_data = sample_data.reshape(1, sample_data.shape[0], sample_data.shape[1])

            # Normalize
            sample_norm = sample_data.reshape(-1, sample_data.shape[-1])
            sample_norm = self.scaler_dl.transform(sample_norm)
            sample_norm = sample_norm.reshape(sample_data.shape)

            prediction = model.predict(sample_norm)
            predicted_class = np.argmax(prediction, axis=1)[0]
            confidence = np.max(prediction)

        else:
            # For classical ML models
            if len(sample_data.shape) == 2:
                sample_data = sample_data.reshape(1, sample_data.shape[0], sample_data.shape[1])

            features = self.extract_features(sample_data)
            features_scaled = self.scaler.transform(features)

            predicted_class = model.predict(features_scaled)[0]
            try:
                probabilities = model.predict_proba(features_scaled)[0]
                confidence = np.max(probabilities)
            except:
                confidence = None

        predicted_activity = self.label_encoder.inverse_transform([predicted_class])[0]

        return predicted_activity, confidence

def main_wisdm(dataset_path):
    """
    Main function specifically for WISDM dataset
    """
    print("Human Activity Recognition Pipeline - WISDM Dataset")
    print("="*60)

    # Initialize the recognizer
    har = HumanActivityRecognizer()

    # Load and preprocess WISDM data
    try:
        X, y = har.load_data(dataset_path=dataset_path)
        X_train, X_val, X_test, y_train, y_val, y_test = har.preprocess_data()
    except Exception as e:
        print(f"Error loading WISDM dataset: {e}")
        print("Falling back to sample data for demonstration...")
        X, y = har.load_data(use_sample_data=True)
        X_train, X_val, X_test, y_train, y_val, y_test = har.preprocess_data()

    # Train classical ML models
    print("\n" + "="*50)
    print("TRAINING CLASSICAL ML MODELS")
    print("="*50)
    classical_models = har.train_classical_models(apply_pca=False)

    # Perform cross-validation
    cv_results = har.cross_validate_classical_models()

    # Train deep learning models
    print("\n" + "="*50)
    print("TRAINING DEEP LEARNING MODELS")
    print("="*50)
    dl_models, history = har.train_deep_learning_models(
        lstm_units=128,  # Increased for better performance on real data
        gru_units=128,
        dropout_rate=0.3,
        batch_size=64,   # Increased batch size for real dataset
        epochs=50        # More epochs for real data
    )

    # Plot training history
    har.plot_training_history()

    # Evaluate all models
    print("\n" + "="*50)
    print("MODEL EVALUATION")
    print("="*50)
    results = har.evaluate_models()

    # Save models
    har.save_models()

    # Compare model performance
    print("\n" + "="*50)
    print("MODEL COMPARISON SUMMARY")
    print("="*50)
    for model_name, result in results.items():
        print(f"{model_name}: {result['accuracy']:.4f}")

    # Find best model
    best_model = max(results.items(), key=lambda x: x[1]['accuracy'])
    print(f"\nBest Model: {best_model[0]} with accuracy {best_model[1]['accuracy']:.4f}")

    # Demonstrate single prediction
    print("\n" + "="*50)
    print("SINGLE PREDICTION EXAMPLE")
    print("="*50)
    sample_idx = 0
    sample_data = X_test[sample_idx]
    actual_activity = har.label_encoder.inverse_transform([y_test[sample_idx]])[0]

    predicted_activity, confidence = har.predict_single_sample(sample_data, best_model[0])

    print(f"Actual Activity: {actual_activity}")
    print(f"Predicted Activity: {predicted_activity}")
    if confidence is not None:
        print(f"Confidence: {confidence:.4f}")

    return har, results

def main():
    """
    Main function to demonstrate the complete pipeline
    """
    print("Human Activity Recognition Pipeline")
    print("="*50)

    # Initialize the recognizer
    har = HumanActivityRecognizer()

    # Load and preprocess data
    X, y = har.load_data(use_sample_data=True)
    X_train, X_val, X_test, y_train, y_val, y_test = har.preprocess_data()

    # Train classical ML models
    print("\n" + "="*50)
    print("TRAINING CLASSICAL ML MODELS")
    print("="*50)
    classical_models = har.train_classical_models(apply_pca=False)

    # Perform cross-validation
    cv_results = har.cross_validate_classical_models()

    # Train deep learning models
    print("\n" + "="*50)
    print("TRAINING DEEP LEARNING MODELS")
    print("="*50)
    dl_models, history = har.train_deep_learning_models(
        lstm_units=64,
        gru_units=64,
        dropout_rate=0.3,
        batch_size=32,
        epochs=30
    )

    # Plot training history
    har.plot_training_history()

    # Evaluate all models
    print("\n" + "="*50)
    print("MODEL EVALUATION")
    print("="*50)
    results = har.evaluate_models()

    # Save models
    har.save_models()

    # Compare model performance
    print("\n" + "="*50)
    print("MODEL COMPARISON SUMMARY")
    print("="*50)
    for model_name, result in results.items():
        print(f"{model_name}: {result['accuracy']:.4f}")

    # Find best model
    best_model = max(results.items(), key=lambda x: x[1]['accuracy'])
    print(f"\nBest Model: {best_model[0]} with accuracy {best_model[1]['accuracy']:.4f}")

    # Demonstrate single prediction
    print("\n" + "="*50)
    print("SINGLE PREDICTION EXAMPLE")
    print("="*50)
    sample_idx = 0
    sample_data = X_test[sample_idx]
    actual_activity = har.label_encoder.inverse_transform([y_test[sample_idx]])[0]

    predicted_activity, confidence = har.predict_single_sample(sample_data, best_model[0])

    print(f"Actual Activity: {actual_activity}")
    print(f"Predicted Activity: {predicted_activity}")
    if confidence is not None:
        print(f"Confidence: {confidence:.4f}")

    return har, results

if __name__ == "__main__":
    # Run the complete pipeline
    recognizer, evaluation_results = main()

    print("\nPipeline completed successfully!")
    print("Models saved and ready for use.")

    # Optional: Hyperparameter tuning example
    # best_rf = recognizer.hyperparameter_tuning('RandomForest')
    # print(f"Best RandomForest model: {best_rf}")
```
